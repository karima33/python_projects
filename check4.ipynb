## Feature selection

```python
corrMatrix = df.corr(method='pearson')
plt.figure(figsize=(16, 16))
sns.heatmap(corrMatrix, cmap="Blues", annot=True)
plt.show()
```

The Pearson correlation coefficient evaluates the strength and direction of the linear relationship between two variables. The coefficient ranges between -1 and +1. The greater the value in absolute term, the stonger is the relationship between two features.



```python
s = corrMatrix.unstack()
so = s.where((s >= 0.50) & (s < 1)).dropna().sort_values(kind="quicksort", ascending=False)
print(so)
```

sqft_above has strong positive correlation with sqft_living and moderate positive correlation with sqft_living15 and sqft_living has positive correlation with sqft_living15. In short there is multi-collinearity issue here, so dropping any 2 columns out of 3.

```python
df.drop('sqft_above', axis=1, inplace=True)
df.drop('sqft_living15', axis=1, inplace=True)
```

When we examine the correlation matrix, we may observe that price has the highest correlation coefficient with living area (sqft_living) (0.7) We will use living area (sqft_living) as feature while creating regression When we model a linear relationship between a response and just one explanatory variable, this is called simple linear regression.



 we notice that "sqft_living",  "grade", "bathrooms" are the higly correlated features with our target variable Price 

# Spliting our dataset into a training set and a testing set

```python
feature_cols =["sqft_living"]
X = df[feature_cols]
y = df.price
```

```python
from sklearn.model_selection import train_test_split
# Splitting the dataset
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
#Import the necessary
from sklearn.linear_model import LinearRegression

# Create/instantiate the linear reg
reg = LinearRegression()

# Train our model
reg.fit(x_train, y_train)

# Check the theta/coef
reg.intercept_,reg.coef_
```

```python
# Predict the x_test with this model and store it in y_pred_skl
y_pred= reg.predict(x_test)
```

```python
#Ploting the linear regression. 

plt.scatter(x_test, y_test, color= 'darkgreen', label = 'data')
plt.plot(x_test, y_pred, color='red', label= ' Predicted Regression line')
plt.xlabel('Living Space (sqft)')
plt.ylabel('price')
plt.legend()
```

```python
#Measuring the performance of linear regression using the testing set

# r2 score 
reg.score(x_test, y_test)
```

The value is low, we can improve the model by including more features and instead of a simple linear regression, we will use multiple linear regression.

# Multiple linear regression

```python
feature_cols =["sqft_living", "grade", "bathrooms"]
X = df[feature_cols]
y = df.price

```

```python
# Splitting the dataset
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
#Import the necessary
from sklearn.linear_model import LinearRegression

# Create/instantiate the linear reg
reg = LinearRegression()

# Train our model
reg.fit(x_train, y_train)

# Check the theta/coef
reg.intercept_,reg.coef_
```

```python
# Predict the x_test with this model and store it in y_pred_skl
y_pred= reg.predict(x_test)
```

```python
# r2 score for multiple linear reg
reg.score(x_test, y_test)
```

we notice that by adding more features and using multiple linear regression  our results  have   slightly improved 

```python
from sklearn.preprocessing import PolynomialFeatures

poly=PolynomialFeatures(degree=2)
lg=LinearRegression()
x_train_fit = poly.fit_transform(x_train) #transforming our input data
lg.fit(x_train_fit, y_train)
x_test_ = poly.fit_transform(x_test)
predicted = lg.predict(x_test_)
```

```python

from sklearn.metrics import mean_squared_error
from sklearn import metrics
print("MSE: ", metrics.mean_squared_error(y_test, predicted))
print("R squared: ", metrics.r2_score(y_test,predicted))
```

we can see that Polynomial has given better results than linear regression and  multiple linear regression and we can improve the r2 score result by adding more features
